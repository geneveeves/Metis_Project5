```python
# Imports
import numpy as np
import pandas as pd
```

```python
# Read in US Foreign Aid dataset
hdi = pd.read_csv('Data/HumanDevelopmentIndex.csv', header=[1]) 
```

```python
hdi.head()
```

```python
# Need to remove all columns with 'Unnamed: _' in the title
hdi = hdi.loc[:, ~hdi.columns.str.contains('^Unnamed')]
```

```python
hdi.info()
```

```python
# Next, convert from HDI values, to change in HDI from previous year
```

```python
hdi.drop(columns = {'HDI Rank (2017)'}, inplace = True)
```

```python
hdi.head()
```

```python
diffs = hdi.diff(axis=1)
```

```python
diffs.head()
```

```python
diffs.Country = hdi.Country
```

```python
diffs.head()
```

```python
# Need to make table have columns for Country, Year, and HDI difference 
# from year before so we can merge with USAID table
diffs = pd.melt(diffs, id_vars = 'Country', 
                value_vars = diffs.columns.values[1:], 
                var_name = 'Year', 
                value_name = 'HDI_Change')
```

```python
diffs.head()
```

```python
diffs.info()
```

```python
# Drop rows with 'NaN' for 'Country'
diffs = diffs[pd.notnull(diffs['Country'])]
```

```python
diffs.info()
```

```python
# Need to convert 'Year' to int
diffs.Year = diffs.Year.apply(lambda x: int(x))
```

```python
diffs.info()
```

```python
# Pickle dataframe
diffs.to_pickle('./hdi_data.pkl')
```
