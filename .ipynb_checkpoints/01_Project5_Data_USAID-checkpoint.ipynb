```python
# Imports
import numpy as np
import pandas as pd
```

```python
# Read in US Foreign Aid dataset
df = pd.read_csv('Data/us_foreign_aid_complete.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
# Drop unnecessary columns
df.drop(columns = {'country_id', 
                   'country_code', 
                   'region_id', 
                   'income_group_id', 
                   'income_group_acronym', 
                   'implementing_agency_id', 
                   'implementing_agency_acronym', 
                   'implementing_subagency_id', 
                   'subagency_acronym', 
                   'channel_category_id', 
                   'channel_subcategory_id', 
                   'channel_id', 
                   'dac_category_id', 
                   'dac_sector_code', 
                   'dac_purpose_code', 
                   'funding_account_id', 
                   'funding_agency_id', 
                   'funding_agency_acronym', 
                   'assistance_category_id', 
                   'aid_type_group_id', 
                   'activity_id', 
                   'activity_project_number', 
                   'activity_start_date', 
                   'activity_end_date', 
                   'transaction_type_id', 
                   'USG_sector_id', 
                   'subagency_name', 
                   'channel_subcategory_name', 
                   'aid_type_group_name', 
                   'submission_id'}, 
       inplace = True)
```

```python
# Check out cleaner dataset
df.head()
```

```python
df.info()
```

```python
# Drop rows with NaN values for 'income_group_name'
df.dropna(subset = ['income_group_name'], inplace = True)
```

```python
# Better!
df.info()
```

```python
# 'fiscal_year' column includes strings. Clean up and make all values integers.
df.fiscal_year.unique()
```

```python
# Looks like there's a weird year entry for '1976tq'
df['fiscal_year'].replace('1976tq', 1976, inplace = True)
```

```python
# Convert all years to integers
df.fiscal_year = df.fiscal_year.apply(lambda x: int(x))
```

```python
df.fiscal_year.unique()
```

```python
df.fiscal_year.max()
```

```python
# Our GDP data only goes up through 2017. Limit data points to 
# those from 1990 - 2017 (more robust data).
df = df.loc[(df['fiscal_year'] >= 1990) & (df['fiscal_year'] <= 2017)]
```

```python
df.info()
```

```python
df.to_pickle('./usaid_data.pkl')
```
