```python
# Imports
import numpy as np
import pandas as pd
```

```python
# Read in US Foreign Aid dataset
df = pd.read_csv('Data/us_foreign_aid_complete.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
# Drop unnecessary columns
df.drop(columns = {'country_id', 
                   'country_code', 
                   'region_id', 
                   'income_group_id', 
                   'income_group_acronym', 
                   'implementing_agency_id', 
                   'implementing_agency_acronym', 
                   'implementing_subagency_id', 
                   'subagency_acronym', 
                   'channel_category_id', 
                   'channel_subcategory_id', 
                   'channel_id', 
                   'dac_category_id', 
                   'dac_sector_code', 
                   'dac_purpose_code', 
                   'funding_account_id', 
                   'funding_agency_id', 
                   'funding_agency_acronym', 
                   'assistance_category_id', 
                   'aid_type_group_id', 
                   'activity_id', 
                   'activity_project_number', 
                   'activity_start_date', 
                   'activity_end_date', 
                   'transaction_type_id', 
                   'USG_sector_id', 
                   'subagency_name', 
                   'channel_subcategory_name', 
                   'aid_type_group_name', 
                   'submission_id'}, 
       inplace = True)
```

```python
df.head()
```

```python
df.info()
```

```python
# Drop rows with NaN values for 'income_group_name'
df.dropna(subset = ['income_group_name'], inplace = True)
```

```python
df.info()
```

```python
# 'fiscal_year' column includes strings. Clean and make all values integers.
df.fiscal_year.unique()
```

```python
# Looks like there's a weird year entry for '1976tq'
df['fiscal_year'].replace('1976tq', 1976, inplace = True)
```

```python
# Convert all years to integers
df.fiscal_year = df.fiscal_year.apply(lambda x: int(x))
```

```python
df.fiscal_year.unique()
```

```python
df.fiscal_year.max()
```

```python
# Limit data points to those since 1990
df = df.loc[df['fiscal_year'] >= 1990]
```

```python
df.info()
```

```python
# Take a look at which countries the US has given the most aid to from 1990-2019
# Dollar amounts are in 2016 USD
df.groupby('country_name').sum().sort_values('constant_amount', ascending = False)['constant_amount']
```

```python
df.groupby('income_group_name').sum().sort_values('constant_amount', ascending = False)['constant_amount']
```

```python
# High Income Countries the US has given aid to since 1990
df.loc[df['income_group_name'] == 'High Income Country']['country_name'].unique()
```

```python
# What type of aid has the US given to High Income Countries? Would assume mostly military
df.loc[df['income_group_name'] == 'High Income Country']['dac_sector_name'].unique()
```
